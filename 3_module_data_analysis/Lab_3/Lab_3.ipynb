### Задание №4*

Удалите выбросы по зарплате (salary) при помощи IQR3 (тройной межквартильный интервал). Сохраните в отдельном датафрейме срез без выбросов.

```python
#slojno
Q1 = df_employee['salary'].quantile(0.25)
Q3 = df_employee['salary'].quantile(0.75)

IQR = Q3 - Q1
low = Q1 - 3 * IQR
up = Q3 + 3 * IQR

df_no_vibros = df_employee[(df_employee['salary'] >= low) & (df_employee['salary'] <= up)]
df_no_vibros.head()
```

### Задание №6

Запишите загруженный датафрейм с доходами в виде таблицы incomes внутри файла sqlite в директории с блокнотом. Не сохраняйте индекс, в случае, если название таблицы уже есть в БД - перепишите эту таблицу.

```python
conn = sqlite3.connect("incomes.db")
sib_fo_income_table.to_sql("incomes", conn, index=False, if_exists="replace")
```

```python
query = "SELECT * FROM incomes"
incomes_df = pd.read_sql(query, conn)
print(incomes_df)
```
